### Location data

```python
# add city to dataframe
city = []
for i, (lat, lon) in df[['latitude', 'longitude']].iterrows():
    if lat<39: # south
        if lon<-110: # southwest
            city.append('Oakland')
        else: # southeast
            city.append('Richmond')
    else: # north
        if lon<-80: # northwest
            city.append('Chicago')
        else: # northeast
            city.append('New Haven')
df['city'] = city
```

```python
cts = df.city.value_counts()

fig, ax = plt.subplots();
ax.bar(range(len(cts)), cts.values);
ax.set_xticks(range(len(cts)))
ax.set_xticklabels(cts.index, rotation=45, size='small');
ax.set_ylabel('Counts');
```

```python
latrange = [24, 50]
lonrange = [-130, -60]
lats = np.arange(24, 50, 5)
lons = np.arange(-130, -60, 10)
m = Basemap(projection='merc', llcrnrlat=latrange[0], urcrnrlat=latrange[1], llcrnrlon=lonrange[0], urcrnrlon=lonrange[1])
m.drawcoastlines()
m.drawmapboundary()
m.drawcountries()
m.drawmeridians(lons)
m.drawparallels(lats)

m.plot(df.longitude, df.latitude, color='r', marker='.', ms=2, ls='', latlon=True);


plt.draw(); plt.show()
```

### Correlations between target variables

```python
%matplotlib inline
fig, axs = plt.subplots(3, 3, figsize = (8, 10))
for i in xrange(len(depvars)):
    for j in xrange(len(depvars)):
        axs[i, j].plot(df['log'+depvars[j]], df['log'+depvars[i]], marker='.', ms=0.75, ls='')
        if i==len(depvars)-1:
            axs[i, j].set_xlabel(depvars[j])
        if j==0:
            axs[i, j].set_ylabel(depvars[i])
```
